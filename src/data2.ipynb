```python
salarios=pd.DataFrame(salaries.groupby('company_location').salary_in_usd.mean())
```

```python
salarios.reset_index(drop=False, inplace=True)
```

```python
salarios.columns
```

```python
paismin=[]
paismed=[]
paismax=[]
for i in range(len(salarios)):
    if salarios.salary_in_usd[i]<33000:
        paismin.append(salarios.company_location[i])
    elif salarios.salary_in_usd[i]>=33000 or salarios.salary_in_usd[i]<73000:
        paismed.append(salarios.company_location[i])
    else:
        paismax.append(salarios.company_location[i])
        
        
```

```python
salaries['codigo_pais']=salaries.salary_in_usd
```

```python

for i in range(len(salaries.company_location)):
    
    if salaries.company_location[i] in paismin:
        
        salaries.codigo_pais[i]=int(1)
        
    elif salaries.company_location[i] in paismed:
        
        salaries.codigo_pais[i]=int(2)
        
    else:
        
        salaries.codigo_pais[i]=int(3)
        
```

```python
paismax
```

```python
salaries['codigo_exp']=salaries.experience_level
```

```python
for i in range(len(salaries.codigo_exp)):
    
    if salaries.experience_level[i]=='SE':
        
        salaries.codigo_exp[i]=int(14)
    elif salaries.experience_level[i]=='MI':
        
        salaries.codigo_exp[i]=int(9)
    elif salaries.experience_level[i]=='EN':
        
        salaries.codigo_exp[i]=int(6)
    else:
        salaries.codigo_exp[i]=int(20)
```

```python
salaries['codigo_remoto']=salaries.remote_ratio
```

```python
for i in range(len(salaries.remote_ratio)):
    
    if salaries.remote_ratio[i]==100:
        
        salaries.codigo_remoto[i]=int(5)
    elif salaries.remote_ratio[i]==0:
        
        salaries.codigo_remoto[i]=int(3)
    elif salaries.remote_ratio[i]==50:
        
        salaries.codigo_remoto[i]=int(50)
    
```

```python
salaries.drop(columns=['company_location','remote_ratio','experience_level','employee_residence','company_size',
                       'salary','salary_currency'],inplace=True)
```
